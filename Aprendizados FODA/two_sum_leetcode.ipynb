## Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

## You may assume that each input would have exactly one solution, and you may not use the same element twice.

## You can return the answer in any order.

Exemplos:
__________________________________
input: ``nums = [2, 7, 11, 15]`` ``target = 9``
output: ``[0, 1]``

___________________________________
input: ``nums = [3, 2, 4]`` ``target = 6``
output: ``[1, 2]``

___________________________________
input: ``nums = [3, 3]``, ``target = 6``
output: ``[0, 1]``

Passos para realizar nesse desafio:

  - criar uma lista ``nums`` com os números a serem adicionados na soma;
  - ``target`` precisa ser, literalmente, ``alvo`` da soma dos numeros;
  - ideia(1): assumindo ``target = x``, ``y`` + ``z`` precisam virar target, percorrendo a lista dada inicialmente. Usar ``map``/``list comprehension`` me parece ideal para isso AGORA.

  - ideia(2): uma lista com os números a serem somados -> cada número precisa somar sobre os outros números uma vez (com exceção dele mesmo, pois não pode ser a soma de dois números iguais) -> fechando o loop após a soma do último número da lista

```python
lista_nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]


lista_numero_somador = [
    numero
    for numero in lista_nums
]

print(lista_numero_somador)
```
