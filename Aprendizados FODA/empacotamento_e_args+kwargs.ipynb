## Empacotamento e desempacotamento de dicionários, além de leve aprofundamento em ``args`` e ``kwargs``

```python
pessoa = {
    'nome': 'lucas',
    'sobrenome': 'marques',
}

# a, b = pessoa.items()

# print(a)
# print(b)

(a1, a2), b = pessoa.items()

# print(a1)
# print(a2)
# print(b)

pessoa_dados = {
    'idade': 15,
    'altura': 1.9
}

pessoa_completa = {**pessoa, 'chave_teste': 'testee', **pessoa_dados}

# print(pessoa_completa)

def mostro_argumentos_nomeados(*args, **kwargs):
  print('Não nomeados: ', args)

  for chave, valor in kwargs.items():
    print(chave, valor)

# mostro_argumentos_nomeados('nao nomeado 1', 'nao nomeado 2', nome='luquinhas', teste=123)

# mostro_argumentos_nomeados(**pessoa_completa)

configuracoes = {
    'arg1': 1,
    'arg2': 2,
    'arg3': 3,
    'arg4': 4,
}

mostro_argumentos_nomeados('nao nomeado rsrsr', 'nao nomeado rsrs 2', **configuracoes)
```
