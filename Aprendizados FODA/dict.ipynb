## Métodos úteis em dicionários no Python
  - len _quantas chaves_
  - keys _iterável com as *chaves*_
  - values _iterável com os *valores*_
  - items _iterável com chaves e valores_
  - setdefault _adiciona valor se a chave não existe_
  - copy (shallow copy) _retorna uma cópia rasa_
  - get _obtém uma chave_ -> útil para evitar exceções
  - pop ~ del _apaga um item com a chave especificada_
  - popitem _apaga o último item adicionado_
  - update _atualiza um dicionário com outro_

```python
## Entendendo melhor a cópia rasa (SHALLOW COPY) ##

d1 = {
    'n1': 1,
    'n2': 2,
    'l1': [0, 1, 2],
}

d2 = d1.copy()

print(d1)
print(d2)

d2['n1'] = 50

print()
print(d1)
print(d2)

d2['l1'][1] = 'vai si fude' ## A mudança na lista ocorre tanto no dicionário d1
                            ## quanto no dicionário d2

print()
print(d1)
print(d2)
```

## Exercício de perguntas e respostas com DICT
- sem muito tratamento de dados

```python
perguntas = [
    {
        'Pergunta': 'Quanto é 2+2?',
        'Opções': ['2', '3', '4', '6'],
        'Resposta': '4',
    },
    {
        'Pergunta': 'Quanto é 5 * 5?',
        'Opções': ['25', '55', '10', '51'],
        'Resposta': '25',
    },
     {
        'Pergunta': 'Quanto é 10 / 2?',
        'Opções': ['3', '5', '7', '1'],
        'Resposta': '5',
    },
]


flag_pergunta = 0
flag_acertos = 0
flag_erros = 0

for item in perguntas:
  print(perguntas[flag_pergunta]['Pergunta'])
  respostaa = input('qual opção é a correta? ')

  if respostaa in item['Opções']:
    print('Acerto miseravi')
    flag_acertos += 1
  else:
    print('errou parceiro busque ajuda')
    flag_erros += 1
  flag_pergunta += 1
print(f'Você acertou {flag_acertos} perguntas e errou {flag_erros} perguntas.')

```
