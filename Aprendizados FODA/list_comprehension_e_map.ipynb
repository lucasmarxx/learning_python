## Compreendendo ``map`` e ``list comprehension``
  > Nota mental:
  - Importante prospectar essa aula para desafios do leetcode!!!!!!!!!

```python
## list comprehension

lista = []
for numero in range(10):
  lista.append(numero)
print(lista)

lista = [
    numero * 2
    for numero in range(10)
]
print(lista)
```

## map - Modificando valores dentro da própria lista

```python
produtos = [
    {'nome': 'p1', 'preco': 20,},
    {'nome': 'p2', 'preco': 10,},
    {'nome': 'p3', 'preco': 30,},
]

novos_produtos = [
    produto['preco'] * 1.1
    for produto in produtos
]

print(*novos_produtos, sep='\n')
```


## Desempacotando em uma nova lista, mudando valores na nova lista

```python
produtos = [
    {'nome': 'p1', 'preco': 20,},
    {'nome': 'p2', 'preco': 10,},
    {'nome': 'p3', 'preco': 30,},
]

novos_produtos = [
    {**produto, 'preco': produto['preco'] * 1.05}
    for produto in produtos
]

print(*novos_produtos, sep='\n')
```

## Mudando valores específicos na lista nova
  > exemplo:
  - apenas UM dos valores e não todos

```python
produtos = [
    {'nome': 'p1', 'preco': 20,},
    {'nome': 'p2', 'preco': 10,},
    {'nome': 'p3', 'preco': 30,},
]

novos_produtos = [
    {**produto, 'preco': produto['preco'] * 1.05}
    if produto['preco'] > 20 else {**produto}
    for produto in produtos
]

print(*novos_produtos, sep='\n')
```
